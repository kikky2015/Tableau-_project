```python
# Count how many rows are missing year, month, membership_type, bicycle_id 
end_station_missing = citibike_df["end_station_name"].isnull().sum()
end_station_id_missing = citibike_df["end_station_id"].isnull().sum()
end_lat_missing= citibike_df["end_lat"].isnull().sum()
end_lng_missing = citibike_df["end_lat"].isnull().sum()

# Create summary table of the missing values 
print(f"There are {end_station_missing} missing end_station values.")
print(f"There are {end_station_id_missing} missing end_station_id values.")
print(f"There are {end_lat_missing} missing end_lat values.")
print(f"There are {end_lng_missing} missing end_lng values.")
```

```python
citibike_df.dropna(inplace=True)
```

```python
citibike_df.count()
```

```python
# Convert the 'started_at' and 'ended_at' to a datetime object
citibike_df['started_at'] = pd.to_datetime(citibike_df['started_at'])
citibike_df['ended_at'] = pd.to_datetime(citibike_df['ended_at'])

# Create separate date and time columns for start date
citibike_df['start_date'] = citibike_df['started_at'].dt.date
citibike_df['start_time'] = citibike_df['started_at'].dt.time

# Create separate date and time columns for end date
citibike_df['end_date'] = citibike_df['ended_at'].dt.date
citibike_df['end_time'] = citibike_df['ended_at'].dt.time

#create day of the week
# Extract the day of the week as a number (0 = Monday, 6 = Sunday)
citibike_df['weekday_number'] = citibike_df['started_at'].dt.weekday

# Map weekday numbers to weekday names
weekday_names = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
citibike_df['trip_day'] = citibike_df['weekday_number'].map(lambda x: weekday_names[x])

```

```python
citibike_df.head()
```

```python
# Drop the original start and end date columns
citibike_df.drop(columns=['started_at'], inplace=True)
citibike_df.drop(columns=['ended_at'], inplace=True)
```

```python
# file path for the CSV file
csv_file_path = 'cleaned_citibike.csv'

# Save the DataFrame to a CSV file
citibike_df.to_csv(csv_file_path, index=False) 
```
